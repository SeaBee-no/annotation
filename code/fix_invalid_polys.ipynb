```python
import geopandas as gpd
from shapely.validation import explain_validity, make_valid
```

```python
shp_path = "../vector/annotation_by_subarea/annotation_KVI.shp"
gdf = gpd.read_file(shp_path)

schema = gpd.io.file.infer_schema(gdf)
for col in gdf.columns:
    dtype = schema["properties"].get(col, None)
    if dtype == "int":
        schema["properties"][col] = "int32:10"
        
gdf["validity"] = gdf.apply(lambda row: explain_validity(row.geometry), axis=1)
gdf
```

```python
# Clean invalid
gdf.geometry = gdf.apply(
    lambda row: make_valid(row.geometry) if not row.geometry.is_valid else row.geometry,
    axis=1,
)

# Check all now valid
gdf["validity"] = gdf.apply(lambda row: explain_validity(row.geometry), axis=1)

gdf
```

```python
del gdf["validity"]
gdf.to_file(f"{shp_path[:-4]}_2.shp", schema=schema)
```
